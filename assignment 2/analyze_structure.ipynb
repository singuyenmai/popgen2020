```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./structure_STR_k2/summary_clustering.tsv", sep="\t")
df.head()
```

```python
ref = pd.read_csv("str_sample_reference", sep="\t")
ref.head()
```

```python
overall = pd.merge(df, ref, on=['ID'], how='inner')
overall.head(15)
```

```python
final = overall.drop(['replication'], axis=1).groupby(by=['ID', 'ref_pop']).mean().reset_index()
final.head()
```

```python
migrant_1 = final.loc[(final['ref_pop']==1) & (final['inferred_cluster_1']<0.9),:]
migrant_1
```

```python
migrant_2 = final.loc[(final['ref_pop']==2) & (final['inferred_cluster_2']<0.9),:]
migrant_2
```

```python
migrant = pd.concat([migrant_1, migrant_2])
migrant
```

```python
overall.to_csv("structure_k2_overall.tsv", sep="\t")
final.to_csv("structure_k2_summary.tsv", sep="\t")
migrant.to_csv("structure_k2_migrant.tsv", sep="\t")
```
